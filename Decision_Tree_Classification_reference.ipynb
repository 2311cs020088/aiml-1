```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
iris = pd.read_csv("iris.csv")
```

```python
iris
```

```python
import seaborn as sns
counts = iris["variety"].value_counts()
sns.barplot(data = counts)
```

```python
iris.info()
```

```python
iris[iris.duplicated(keep = False)]
```

#### Observation
- There is no null values
- Y-columns are the "variety" which is categorical
- x-columns are the sepal-length,width,length
- petal-length,width
- There is no duplicated rows
- There are 150 rows and 5 columns
- There are three flower categories(classes)
- All x-columns are continuous

```python
iris = iris.drop_duplicates(keep = 'first')
iris[iris.duplicated]
```

```python
iris = iris.reset_index(drop=True)
iris
```

```python
iris["variety"].value_counts()
```

```python
labelencoder = LabelEncoder()
iris.iloc[:, -1] = labelencoder.fit_transform(iris.iloc[:,-1])
iris.head()
```

```python
iris.info()
```

#### Observation
- The target column('variety') is still object type. it needs to be converted to numeric(int)

```python
iris['variety'] = pd.to_numeric(labelencoder.fit_transform(iris['variety']))
print(iris.info())
```

```python
iris.head(4)
```

```python
iris.head(3)
X = iris.iloc[:,0:4]
Y = iris['variety']
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3,random_state = 1)
x_train.head(20)
```

```python
x_train
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth= None)
model.fit(x_train,y_train)
```

```python
plt.figure(dpi=1200)
tree.plot_tree(model);
```

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
plt.figure(dpi = 1200)
tree.plot_tree(model,feature_names=fn,class_names=cn)
```

```python
preds = model.predict(x_test)
preds
```

```python
print(classification_report(y_test,preds))
```
